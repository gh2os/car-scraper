```python
import sqlite3
import pandas as pd
from datetime import datetime

# Connect to your local listings database
conn = sqlite3.connect("data/listingsnew.db")
```

```python
query = """
SELECT 
    l.id,
    l.title,
    l.location,
    l.link,
    l.first_seen,
    l.last_seen,
    l.is_delisted,
    s.price,
    s.mileage
FROM listings l
LEFT JOIN (
    SELECT listing_id, MAX(scraped_at) AS latest_scrape
    FROM listing_snapshots
    GROUP BY listing_id
) latest ON l.id = latest.listing_id
LEFT JOIN listing_snapshots s
    ON s.listing_id = latest.listing_id AND s.scraped_at = latest.latest_scrape
"""

df = pd.read_sql_query(query, conn)
```

```python
df["first_seen"] = pd.to_datetime(df["first_seen"])
df["last_seen"] = pd.to_datetime(df["last_seen"])
df["days_on_market"] = (df["last_seen"] - df["first_seen"]).dt.total_seconds() / 86400
```

```python
df["first_seen"] = pd.to_datetime(df["first_seen"])
df["last_seen"] = pd.to_datetime(df["last_seen"])

# Time on market in days
df["days_on_market"] = (df["last_seen"] - df["first_seen"]).dt.total_seconds() / 86400

df[["title", "location", "price", "days_on_market"]].head()
```

```python
import re

def parse_title(title):
    if not isinstance(title, str):
        return None, None, None
    match = re.match(r"^(20[0-2][0-9])\s+(\w+)\s+([\w\-]+)", title)
    if match:
        return match.group(1), match.group(2), match.group(3)
    return None, None, None

df[["model_year", "make", "model"]] = df["title"].apply(lambda x: pd.Series(parse_title(x)))
df["model_year"] = pd.to_numeric(df["model_year"], errors="coerce")
df.head()
```

```python
cx5_df = df[df["make"] == "Mazda"].copy()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 5))
sns.histplot(cx5_df["price"], bins=30, kde=True)
plt.title("Mazda CX-5 Price Distribution")
plt.xlabel("Price")
plt.ylabel("Listings")
plt.grid(True)
plt.show()
```

```python
yearly = cx5_df.groupby("model_year")["price"].agg(["count", "mean", "min", "max"]).round(0)
display(yearly)
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=cx5_df, x="mileage", y="price", hue="model_year", palette="viridis", alpha=0.6)
plt.title("Mazda CX-5: Price vs Mileage")
plt.xlabel("Mileage (km)")
plt.ylabel("Price ($)")
plt.grid(True)
plt.show()
```

```python
price_mileage_summary = (
    cx5_df.groupby("model_year")[["price", "mileage"]]
    .agg(["count", "mean", "min", "max"])
    .round(0)
    .astype(int)
)

# Rename columns for clarity
price_mileage_summary.columns = [
    "price_count", "price_avg", "price_min", "price_max",
    "mileage_count", "mileage_avg", "mileage_min", "mileage_max"
]

display(price_mileage_summary)
```

```python
cx5_df["price_per_km"] = (cx5_df["price"] / cx5_df["mileage"]).round(2)
```

```python
mean_ppk = cx5_df["price_per_km"].mean()
std_ppk = cx5_df["price_per_km"].std()

cx5_df["value_score"] = ((cx5_df["price_per_km"] - mean_ppk) / std_ppk).round(2)
```

```python
best_value = cx5_df.sort_values("value_score").head(30)
display(best_value[["model_year", "title", "mileage", "price", "price_per_km", "value_score", "location"]])
```
